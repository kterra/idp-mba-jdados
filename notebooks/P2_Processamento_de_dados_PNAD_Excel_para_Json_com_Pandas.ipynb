# Processamento de dados PNAD 

Fonte dos dados: https://sidra.ibge.gov.br/tabela/7526#resultado

```python
import pandas as pd
import json
```

```python
#substitua esse caminho pelo caminho do arquivo no seu Drive
PNAD_FILE_PATH ='/content/drive/MyDrive/Colab Datasets/Tabela-7526-2020.xlsx'
```

```python
pnad_renda_2020 = pd.read_excel(PNAD_FILE_PATH , skiprows=[0,1,2,3,17])
```

```python
pnad_renda_2020
```

```python
pnad_renda_2020.columns = ['from','lowerLimit']
pnad_renda_2020['from'] = pnad_renda_2020['from'].str[1:].astype('int')
```

```python
pnad_renda_2020_to = pnad_renda_2020['from'].shift(-1, fill_value=100).astype('int')

pnad_renda_2020.insert(1, 'to', pnad_renda_2020_to) 
```

```python
pnad_renda_2020
```

```python
result = pnad_renda_2020.to_json(orient='records')
parsed = json.loads(result)
json_object = json.dumps(parsed, indent=4)
```

```python
with open("pnad-renda-2020.json", "w") as outfile:
    outfile.write(json_object)
```
