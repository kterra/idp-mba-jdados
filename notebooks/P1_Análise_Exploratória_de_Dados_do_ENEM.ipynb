## Medidas descritivas

**Moda:** O valor mais frequente. Normalmente utilizada para variáveis nominais.

**Média aritmética simples:** soma dos valores divida pelo número de elementos.

**Mediana:** é o valor central do conjunto de valores quando ordenados de forma crescente. Em caso de número par, será a média dos dois valores centrais da amostra.

**Variância:** é a distância quadrática média em relação à média. E[(X-E(X))²]

Desvio-padrão: é uma medida de dispersão em torno da média da variável. Medida pela raiz quadrada da variância.

### Frequência e Distribuição

1. Histograma: é uma representação da função de distribuição de probabilidades. (b.h=probabilidade|f(x).x=p(x))
1. Gráfico de frequência absoluta ou relativa: indica a distribuição das variáveis.
1. Relação entre variávies


```python
microdados_enem_2019['TP_COR_RACA'].value_counts(normalize=True).sort_index()
```

```python
microdados_enem_2019['TP_COR_RACA'].value_counts(normalize=True).sort_index().plot(kind='bar')
```

```python
microdados_enem_2019['TP_SEXO'].value_counts(normalize=True).sort_index().plot(kind='bar')
```

```python
#Calcular MEDIA
microdados_enem_2019['MEDIA'] = (microdados_enem_2019['NU_NOTA_CH'] + \
                                 microdados_enem_2019['NU_NOTA_CN'] + \
                                 microdados_enem_2019['NU_NOTA_LC'] + \
                                  microdados_enem_2019['NU_NOTA_MT']+\
                                  microdados_enem_2019['NU_NOTA_REDACAO'])/5
```

```python
microdados_enem_2019['MEDIA'].plot.hist(bins=60, figsize=(15,5))
```

```python
microdados_enem_2019.Q006.value_counts().sort_index()
```

```python
df_q006_A = microdados_enem_2019[microdados_enem_2019.Q006 == 'A']
df_q006_B = microdados_enem_2019[microdados_enem_2019.Q006 == 'B']
df_q006_Q = microdados_enem_2019[microdados_enem_2019.Q006 == 'Q']
```

```python
sns.distplot(df_q006_A.MEDIA)
sns.distplot(df_q006_B.MEDIA)
sns.distplot(df_q006_Q.MEDIA)
```

```python
#gerando todos os gráficos
classes_renda = microdados_enem_2019.Q006.unique()
for classe in classes_renda:
  df_Q006_classe = microdados_enem_2019[microdados_enem_2019.Q006 == classe]
  sns.distplot(df_Q006_classe.MEDIA)
```
